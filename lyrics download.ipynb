```python
from time import time
```

```python
def dl_lyrics(table):
    start_time = time()
    lyrics = []
    for i in range(len(table)):
        try:
            lyrics.append(PyLyrics.getLyrics(table.Artist[i], table.Song[i]))
        except:
            lyrics.append('not available')
    print(time() - start_time)
    print(len(lyrics))
    table['lyrics'] = lyrics
    return table
```

```python
start_time = time()
lyrics = []
for i in range(len(max_martin)):
    try:
        lyrics.append(PyLyrics.getLyrics(max_martin.Artist[i], max_martin.Song[i]))
    except:
        lyrics.append('not available')

len(lyrics)
```

```python
time()-start_time
```

```python
max_martin['lyrics'] = lyrics
```
